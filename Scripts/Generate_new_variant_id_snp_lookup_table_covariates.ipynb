Add a new variant_id and keep the old variant id(rs id) which added by gatk

```python
df1=mygenotype.copy()
df2=mygenotype[['#CHROM','POS','ID','REF','ALT']]
df2.assign(ID_old=None)
for index,row in df1.iterrows():
    id_o=row['ID']
    df1.at[index,'ID']=row['#CHROM']+'_'+str(row['POS'])+'_'+row['REF']+'_'+row['ALT']
    df2.at[index,'ID']=row['#CHROM']+'_'+str(row['POS'])+'_'+row['REF']+'_'+row['ALT']
    df2.at[index,'ID_old']=id_o   
```

```python
df2.rename(columns={'#CHROM':'chr','POS':'variant_pos','ID':'variant_id','ID_old':'rs_id_dbsnp'},inplace = True)
```
